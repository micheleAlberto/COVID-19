```python
import pandas as pd
```

```python
confirmed_df = pd.read_csv("csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
interesting_countries = [
    "Italy", "Greece", "Turkey", "France", "Netherlands", "Spain", "United Kingdom", "Portugal"
]
last_month_columns = [c for c in confirmed_df.columns if len(c.split("/"))==3][-31:]
```

```python
confirmed_selection = confirmed_df[
    confirmed_df["Country/Region"].map(lambda c: c in interesting_countries)
][
    confirmed_df["Province/State"].isna()
][["Country/Region"]+last_month_columns]
confirmed_selection
```

```python
def difference(df):
    baseline = df[last_month_columns[0]] 
    new_df= pd.DataFrame({
        "Country/Region": df["Country/Region"]
    })
    for date in last_month_columns:
        new_df[date]= df[date] - baseline
    return new_df
    
def ratio(df):
    baseline = df[last_month_columns[0]] 
    new_df= pd.DataFrame({
        "Country/Region": df["Country/Region"]
    })
    for date in last_month_columns:
        new_df[date]= (df[date] / baseline) - 1.
    return new_df

def flip(df):
    flipped = df.transpose().reset_index()
    flipped.columns = flipped.iloc[0]
    flipped = flipped.iloc[1:]
    return flipped
```

```python
import plotly.express as px
for c in interesting_countries:
    fig = px.line(flipped, x='Country/Region', y=c, labels=c)
```

```python
import plotly.graph_objs as go
def draw(df):
    fig = go.Figure()
    for c in interesting_countries:
        fig.add_trace(go.Scatter(
            x=df['Country/Region'], 
            y=df[c],
            name=c,
            mode='lines'))
    fig.show()
```

# Normalize by population

```python
def normalize_by_population(df):
    population = pd.read_csv("csse_covid_19_data/UID_ISO_FIPS_LookUp_Table.csv")
    population = population[
        population["Country_Region"].map(lambda c: c in interesting_countries)
    ][
        population["Province_State"].isna()
    ][["Country_Region","Population"]]
    merged = pd.merge(df, population, left_on="Country/Region", right_on="Country_Region")
    for d in last_month_columns:
        merged[d] = merged[d] / merged["Population"]
    return merged[["Country/Region"]+last_month_columns]
```

```python
flip(normalize_by_population(confirmed_selection))
```

```python
draw(flip(normalize_by_population(confirmed_selection)))
```

# Population Normalized increments

```python
(flip(normalize_by_population(difference(confirmed_selection))))
```

```python
draw(flip(normalize_by_population(difference(confirmed_selection))))
```
